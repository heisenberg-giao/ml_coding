```python
from heapq import heappush, heappop, heapify
def getOrder(tasks):
    """
    1. add the index into the tasks list
    2. sort the tasks list based on enque time, as it determines whether the task can be enqued compared with current time T. 
    3. push the available task that can be enqued into the heap
    4. Inside the minHeap, pop the task with min processing time and record the index.
    """
    # [[1,2],[2,4],[3,2],[4,1]]
    for i, t in enumerate(tasks): 
        t.append(i)
    # [[1,2,0],[2,4,1],[3,2,2],[4,1,3]]

    # sort tasks based on enque time
    tasks.sort(key = lambda x: x[0])

    # record index and current time (enque time for first element)
    i, T = 0, tasks[0][0]
    res, minHeap = [], []
    
    # while heap not empty or tasks have elements left
    while minHeap or i<len(tasks):
        while i<len(tasks) and T>=tasks[i][0]:
            heappush(minHeap, [tasks[i][1], tasks[i][2]]) # we are only interested in processing time and index; and the minheap's top is the task with min processing time and smaller index.
            i += 1
        if not minHeap: # CPU is idle, we should fast forward to the next enque time in the tasks list.
        # need to fast forwarding instead of incrementing by 1 only 
            T = tasks[i][0]
        else:
            procTime, index = heappop(minHeap)
            T += procTime
            res.append(index)
    return res
    # both sorting and the heap pop/push is O(nlogn).
    # space: O(N) for minheap
                


        
```

"""Applications
给定一个映射 freq，其中 freq[x] 表示元素 x 的出现次数。请完成以下两个任务：
判断是否存在排列：是否能将所有元素排列成一个序列，使得相邻的两个元素不相同？
构造排列（如果存在）：返回任意一个满足条件的排列。
Input: freq = {'a': 2, 'b': 2}
Output: True, ['a', 'b', 'a', 'b']  # 或 ['b', 'a', 'b', 'a']

Input: freq = {'a': 3, 'b': 1}
Output: False, None  # 无法排列，因为 'a' 过多

Input: freq = {'a': 3, 'b': 2, 'c': 1}
Output: True, ['a', 'b', 'a', 'b', 'a', 'c']  # 其他合法排列也可

"""



```python
def can_rearrange(freq):
    maxCount = max(freq.values())
    maxChar = max(freq.keys(), key=lambda k: freq[k])
    totalCount = sum(freq.values())

    # 1. Check if the most frequent char can be placed without adjacent duplicates
    if maxCount > (totalCount+1) // 2:
        return False, None 
    
    
    res = [None] * totalCount
    i = 0
    while freq[maxChar] > 0:
        res[i] = maxChar 
        i += 2
        freq[maxChar] -= 1
    
    for char, count in freq.items():
        while count > 0:
            if i>=totalCount:
                i = 1
            res[i] = char 
            i += 2
            count -= 1
    return True, res



freq = {'a': 3, 'b': 2, 'c': 1}
freq = {'a': 2, 'b': 2}

can_rearrange(freq)
```

输出: ["(a())()", "(a)()()"]
输入: s = "()())()"
输出: ["()()()", "(())()"]

```python
from collections import deque
def removeInvalidParentheses(s):
    res = []
    visited = set()
    q = deque([s])
    # 一旦找到合法解，只处理这一层的所有解，不再向下删括号
    found = False 

    def is_valid(string):
        count = 0
        for c in string:
            if c == '(':
                count += 1
            elif c == ')':
                count -= 1
            if count < 0:
                return False
        return count == 0
    
    while q:
        curr = q.popleft()
        if is_valid(curr):
            res.append(curr)
            found = True 
        # 如果找到了合法解，就只在当前这一层处理所有结果就行，不能再向下删除了，因为事要求
        if found:
            continue 
        for i in range(len(curr)):
            if curr[i] not in ('(', ')'):
                continue 
            next_str = curr[:i] + curr[i+1:]
            if next_str not in visited:
                visited.add(next_str)
                q.append(next_str)
    return list(set(res))

removeInvalidParentheses('(a)())()')
```

```python
def generateParethesis(n):
    res = []

    def is_valid(string):
        count = 0
        for c in string:
            if c == '(':
                count += 1
            elif c == ')':
                count -= 1
            if count < 0:
                return False
        return count == 0
    
    def backtrack(curr):
        if len(curr) == 2*n:
            if is_valid(curr):
                res.append(''.join(curr))
                return
        else:
            curr.append('(')
            backtrack(curr)
            curr.pop()

            curr.append(')')
            backtrack(curr)
            curr.pop()
    backtrack([])
    return res 

generateParethesis(3)
    
```

```python
def wordSearch(board, word):
    ROWS, COLS = len(board), len(board[0])
    visited = set()

    def backtrack(r, c, i):
        """
        r, c is the current location on the board
        i is the ith index in the word
        """
        # stopping
        if i == len(word):
            return True 
        # conditions that we need to stop
        if r<0 or c<0 or r>=ROWS or c>=COLS or board[r][c] != word[i] or (r, c) in visited:
            return False 
        
        # start the next step
        visited.add((r, c))
        res = (
        backtrack(r+1, c, i+1) or
        backtrack(r-1, c, i+1) or
        backtrack(r, c+1, i+1) or
        backtrack(r, c-1, i+1) 
        )
        visited.remove((r, c))
        return res 
    
    for c in range(COLS):
        for r in range(ROWS):
            if backtrack(r, c, 0):
                return True 
    return False

board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]
word = "ABCCED"
wordSearch(board, word)
```

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.isWord = False

class Trie:
    def __init__(self):
        self.root = TrieNode()
    
    def insert(self, word):
        node = self.root
        for c in word:
            if c not in node.children:
                node.children[c] = TrieNode()
            node = node.children[c]
        node.isWord = True 

    def search(self, word):
        node = self.root
        for c in word:
            if c not in node.children:
                return False 
            node = node.children[c]
        return node.isWord



def WordSearchII(board, words):
    trie = Trie()
    for word in words:
        trie.insert(word)
    
    ROWS, COLS = len(board), len(board[0])
    res, visited = set(), set()

    def dfs(r, c, node, word):
        if node.isWord:
            res.add(word)
            # once found the word, label it as False
            node.isWord = False 
        if r<0 or c<0 or r>=ROWS or c>=COLS or (r, c) in visited or board[r][c] not in node.children:
            return 
        
        # do
        visited.add((r,c))
        word += board[r][c]

        curr = node.children[board[r][c]]

        # backtrack
        dfs(r+1, c, curr, word)
        dfs(r-1, c, curr, word)
        dfs(r, c+1, curr, word)
        dfs(r, c-1, curr, word)

        # undo
        visited.remove((r,c))
        # we don't need to undo the word since string is immutable but list or set is mutable

    for r in range(ROWS):
        for c in range(COLS):
            dfs(r, c, trie.root, "")
    return list(res)

board = [["o","a","a","n"],["e","t","a","e"],["i","h","k","r"],["i","f","l","v"]]
words = ["oath","pea","eat","rain"]
WordSearchII(board, words)
```

```python
# two follow ups
# 1. variant: instead of outputting all strings, also ask to output the path


class TrieNode:
    def __init__(self):
        self.children = {}
        self.word = None

class Trie:
    def __init__(self):
        self.root = TrieNode()
    
    def insert(self, word):
        node = self.root
        for c in word:
            if c not in node.children:
                node.children[c] = TrieNode()
            node = node.children[c]
        node.word = word 



def WordSearchII_with_paths(board, words):
    trie = Trie()
    for word in words:
        trie.insert(word)
    
    ROWS, COLS = len(board), len(board[0])
    res, visited = {}, set() # instead of store a set, store a dictionary for the path
    # {word: [(r1,c1), (r2,c2),...]}

    # dfs takes in current location (r, c), current node in the trie, and the current path that have formulated
    def dfs(r, c, node, path):
        if node.word:
            res[node.word] = path.copy()
            # once found the word, label it as None
            node.word = None 
        if r<0 or c<0 or r>=ROWS or c>=COLS or (r, c) in visited or board[r][c] not in node.children:
            return 
        
        # do
        visited.add((r,c))
        path.append((r,c))
        char = board[r][c]
        next_node = node.children[char]

        # backtrack
        dfs(r+1, c, next_node, path)
        dfs(r-1, c, next_node, path)
        dfs(r, c+1, next_node, path)
        dfs(r, c-1, next_node, path)

        # undo
        path.pop()
        visited.remove((r,c))

    for r in range(ROWS):
        for c in range(COLS):
            dfs(r, c, trie.root, [])
    return res
board = [["a","b","c"],
         ["d","e","f"],
         ["g","h","i"]]

words = ["abc", "cfi", "beh", "defi", "gh"]
WordSearchII_with_paths(board, words)
```

```python
# 2. 不同的words之间不能reuse同一个board cell，求：符合要求的size的最大结合
def max_non_overlapping_words(board, words):
    # 1. get all possible words and the path in the board
    # return {'oath': [(0, 0), (0, 1), (1, 1), (2, 1)], 'eat': [(1, 3), (1, 2), (1, 1)]}
    word_paths = WordSearchII_with_paths(board, words)
    # 2. sort by greedy, shorter words first and easier to exist with other words
    words_sorted = sorted(word_paths.keys(), key=lambda x: len(x))

    # 3. init res and occupied (r,c)
    selected = [] # store final selected non-overlapping words
    occupied = set() # record (r,c) that is occupied

    # 4. loop through the words list
    for word in words_sorted:
        path = word_paths[word]

        conflict = False
        for coordinate in path:
            if coordinate in occupied:
                conflict = True
                break
        if not conflict:
            selected.append(word)
            occupied.update(path)
    return selected

board = [["a","b","c"],
        ["d","e","f"],
        ["g","h","i"]]

words = ["abc", "cfi", "beh", "defi", "gh"]
max_non_overlapping_words(board, words)
```

```python
from heapq import heapify, heappush, heappop
class MedianFinder:

    def __init__(self):
        self.left, self.right = [], []
        

    def addNum(self, num: int) -> None:
        heappush(self.left, -num)
        heappush(self.right, -heappop(self.left))
        if len(self.left)<len(self.right):
            heappush(self.left, -heappop(self.right))

    def findMedian(self) -> float:
        if len(self.left) > len(self.right):
            return -self.left[0]
        else:
            return (-self.left[0]+self.right[0])/2

# Your MedianFinder object will be instantiated and called as such:
obj = MedianFinder()
obj.addNum(1)
obj.addNum(3)
print(obj.findMedian())
obj.addNum(5)
print(obj.findMedian())
```

```python
class MedianFinder:

    def __init__(self):
        self.freq = [0] * 101
        self.total = 0

    def addNum(self, num: int) -> None:
        self.freq[num] += 1
        self.total += 1
        

    def findMedian(self) -> float:
        count = 0 # count the number of elements from small to high
        # 为了找到中位数，odd的时候时候是中位数和中位数下一位；even的时候是两个有效中位数
        m1, m2 = None, None
        for i in range(101):
            count += self.freq[i]
            # m1只赋值一次，寻找第（N+1)/2 下界的个数，
            # # (self.total+1)//2: 求的是第几个数，
            # 一共5个数，求的就是第3个，下面求的是第四个，但是最后return不会用
            # 一共6个数，求的就是第3个，下面求的是第4个；
            if not m1 and count >= (self.total+1)//2:
                m1 = i 
            # 寻找第（N+2)/2 个数
            # 一旦找到两个位置，就不用遍历了
            if count >= (self.total+2)//2:
                m2 = i 
                break 
        if self.total%2 == 1:
            return m1
        else:
            return (m1+m2)/2
# Your MedianFinder object will be instantiated and called as such:
obj = MedianFinder()
obj.addNum(1)
obj.addNum(3)
print(obj.findMedian())
obj.addNum(5)
obj.addNum(5)
obj.addNum(9)
obj.addNum(10)
obj.addNum(11)
obj.addNum(12)

print(obj.findMedian())
```
